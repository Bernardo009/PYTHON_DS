```python
import pandas as pd

# dataframe

data = {
    "Nome": ["Pedro", "Maria", "João", "Ana", "Thiago"],
    "Idade": [22, 15, 33, 19, 28],
    "Cargo": ["Dev", "Gerente", "Ceo", "Dev", "Diretor"],
    "Salario": [3800, 5200, 12000, 3800, 10000],
}

df = pd.DataFrame(data)

print(df)
```

```python
import pandas as pd

# métodos e atributos

# max - 

print(df["valores"].max())

# idxmax - 

print(df["valores"].idmax())

# head

print(df.head(2))

# tail - faz a mesma coisa que o head, mas pega as infomações de baixo para cima

print(df.tail(2))

# info

df.info()

# dtypes

print(df.dtypes)
```

```python
import pandas as pd
# mean

dados_dict = {
    "aluno": ["Ana", "Bruno", "Calos", "Daniel", "Eduardo"],
    "nota": [8.5, 7.5, 5, 9, 7.5]
}

dataF = pd.DataFrame(dados_dict)

print(dataF["nota"].mean())
```

```python
import pandas as pd
# sort_values - coordenação dos valores

dados_dict = {
    "aluno": ["Ana", "Bruno", "Calos", "Daniel", "Eduardo"],
    "nota": [8.5, 7.5, 5, 9, 7.5]
}

dataF = pd.DataFrame(dados_dict)

print(dataF.sort_values(by="nota", ascending=True))
```
