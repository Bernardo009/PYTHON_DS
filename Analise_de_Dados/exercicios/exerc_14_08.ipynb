### Realize a criação de um grafico de linhas para exibir os dados apresentados abaixo


```python
import matplotlib.pyplot as plt

ano = [
    2009,
    2010,
    2011,
    2012,
    2013,
    2014,
    2015,
    2016,
    2017,
    2018,
    2019,
    2020,
    2021,
    2022,
]

receita = [
    45.9,
    65.2,
    108.2,
    156.5,
    170.9,
    182.8,
    233.7,
    215.6,
    229.2,
    265.6,
    260.2,
    274.5,
    365.8,
    394.3,
]

plt.plot(ano, receita, marker="o")

plt.grid(True)

plt.xlabel("Ano")
plt.ylabel("Receita")

plt.title("Correlação entre ano e receita")
plt.show()
```

### Realizar a criação de um grafico de barras e um grafico de dispersão com os dados de ano e receita


```python
ano = [
    2009,
    2010,
    2011,
    2012,
    2013,
    2014,
    2015,
    2016,
    2017,
    2018,
    2019,
    2020,
    2021,
    2022,
]

receita = [
    45.9,
    65.2,
    108.2,
    156.5,
    170.9,
    182.8,
    233.7,
    215.6,
    229.2,
    265.6,
    260.2,
    274.5,
    365.8,
    394.3,
]

plt.scatter(ano, receita)

plt.xlabel("Ano")
plt.ylabel("Receita")

plt.title("Correlação entre ano e receita")
plt.grid(True)

plt.show()
```

```python
ano = [
    2009,
    2010,
    2011,
    2012,
    2013,
    2014,
    2015,
    2016,
    2017,
    2018,
    2019,
    2020,
    2021,
    2022,
]

receita = [
    45.9,
    65.2,
    108.2,
    156.5,
    170.9,
    182.8,
    233.7,
    215.6,
    229.2,
    265.6,
    260.2,
    274.5,
    365.8,
    394.3,
]

bars = plt.bar(ano, receita)

plt.bar_label(bars, labels=receita, padding=2, fontsize=8, fontweight="bold")

plt.xlabel("Ano")
plt.ylabel("Receita")

plt.title("Correlação entre ano e receita")

plt.show()
```

### Realizar a criação de um gráfico de pizza com os dados abaixo

```python
fabricante = ["Samsung", "Apple", "Huawei", "Xiaomi", "Oppo", "Outros"]

participacao = [30.2, 26.6, 10.4, 9.7, 4.8, 18.3]

cores = plt.get_cmap("tab20c").colors

plt.pie(participacao, labels=fabricante, autopct="%.1f", colors=cores)

plt.title("Percentual")

plt.show()
```
